# **Insertion Sort**

**Algoritma Insertion sort**

```python
#Insertion sort
#sort secara ascending
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        temp = arr[i]
        j = i - 1
        while j >= 0 and temp < arr[j]:
            arr[j + 1] = arr[j]
            j =j - 1
        arr[j + 1] = temp

arr = [7,1,6,2,8,3,9,10]

#best case
print("best case")
arr1 = [1,2,3,4,5,6,7,9,8]
print("Array sebelum sort: ",arr1)
insertion_sort(arr1)
print("Array setelah sort: ",arr1)

#average case
print("average case")
arr2=[5, 2, 1, 4, 3]
print("Array sebelum sort: ",arr2)
insertion_sort(arr2)
print("Array setelah sort: ",arr2)

#worst case
print("worst case")
arr3=[5, 2, 1, 4, 3]
print("Array sebelum sort: ",arr3)
insertion_sort(arr3)
print("Array setelah sort: ",arr3)
```

**Cara kerja algoritma**

Inisialisasi Variabel:

Inisialisasi variabel i ke 1 (indeks pertama setelah elemen pertama).
Inisialisasi variabel temp.

**Perulangan**:

Masuk ke dalam perulangan for dengan iterasi i dari 1 hingga len(arr).
Set temp sebagai elemen pada indeks i.
Inisialisasi variabel j sebagai indeks sebelum i.

**Pembandingan dan Pergeseran**:

Selama j lebih dari atau sama dengan 0 dan elemen pada indeks j lebih besar dari temp, lakukan pergeseran elemen ke posisi berikutnya.
Kurangi j.
Penempatan Elemen:
Setelah keluar dari perulangan, tempatkan key pada posisi j + 1.

**Hasil**:

Setelah iterasi selesai, arr akan terurut.

**Selesai**

**Algoritma selesai**.

# **Bubble sort**
**Algoritma bubble sort**

```python
#Bubble sort ascending
def bubble_sort(arrInt):
    n = len(arrInt)
    i = 0
    while i < n - 1:
        j = 0
        while j < n - i - 1:
            if arrInt[j] > arrInt[j + 1]:
                temp = arrInt[j]
                arrInt[j] = arrInt[j + 1]
                arrInt[j + 1] = temp
            j+=1
        i+=1

#best case
print("best case")
arr1 = [1,2,3,4,5,6,7,9,8]
print("Array sebelum sort: ",arr1)
bubble_sort(arr1)
print("Array setelah sort: ",arr1)

#average case
print("average case")
arr2=[5, 2, 1, 4, 3]
print("Array sebelum sort: ",arr2)
bubble_sort(arr2)
print("Array setelah sort: ",arr2)

#worst case
print("worst case")
arr3=[5, 2, 1, 4, 3]
print("Array sebelum sort: ",arr3)
bubble_sort(arr3)
print("Array setelah sort: ",arr3)
```

**Cara kerja algoritma**

Inisialisasi Variabel:

Inisialisasi variabel n dengan panjang array arrInt.

Inisialisasi variabel i dengan 0.

**Perulangan Luar**:

Masuk ke dalam perulangan while dengan kondisi i < n - 1.

Inisialisasi variabel j dengan 0.

**Perulangan Dalam**:

Masuk ke dalam perulangan while dengan kondisi j < n - i - 1.

Pembandingan elemen: Periksa apakah elemen pada indeks j lebih besar
dari elemen pada indeks j + 1.
Jika ya, lakukan pertukaran elemen menggunakan variabel sementara temp.

**Pembaruan Indeks:**

Tambahkan j sebanyak 1 setelah keluar dari perulangan dalam.

**Pembaruan Indeks luar:**

Tambahkan i sebanyak 1 setelah keluar dari perulangan luar.

**Selesai**

Algoritma selesai.

**Perbandingan efektivitas waktu antara bubble sort dan insertion sort**

```python
import time
import random
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        temp = arr[i]
        j = i - 1
        while j >= 0 and temp < arr[j]:
            arr[j + 1] = arr[j]
            j =j - 1
        arr[j + 1] = temp
def bubble_sort(arrInt):
    n = len(arrInt)
    i = 0
    while i < n - 1:
        j = 0
        while j < n - i - 1:
            if arrInt[j] > arrInt[j + 1]:
                temp = arrInt[j]
                arrInt[j] = arrInt[j + 1]
                arrInt[j + 1] = temp
            j+=1
        i+=1

def generate_random_data(size):
    return [random.randint(1, 10000) for _ in range(size)]

def measure_time(sort_function, data):
    start_time = time.time()
    sort_function(data)
    end_time = time.time()
    return end_time - start_time



data_sizes = [10, 100, 1000, 5000,10000]

insertion_times = []
bubble_times = []

for size in data_sizes:
    data_test = generate_random_data(size)

    insertion_time = measure_time(insertion_sort, data_test.copy())
    bubble_time = measure_time(bubble_sort, data_test.copy())

    insertion_times.append(insertion_time)
    bubble_times.append(bubble_time)

# Menampilkan waktu eksekusi
for i, size in enumerate(data_sizes):
    print(f"Waktu eksekusi untuk {size} data:")
    print(f"Insertion Sort: {insertion_times[i]:.6f} detik")
    print(f"Bubble Sort: {bubble_times[i]:.6f} detik")
    print()

# Menampilkan grafik line chart
plt.figure(figsize=(10, 6))
plt.plot(data_sizes, insertion_times, marker='o', label='Insertion Sort', color='skyblue')
plt.plot(data_sizes, bubble_times, marker='o', label='Bubble Sort', color='salmon')
plt.xlabel('Jumlah Data')
plt.ylabel('Waktu Eksekusi (detik)')
plt.xticks(data_sizes)
plt.title('Perbandingan Waktu Eksekusi Insertion Sort dan Bubble Sort pada Data Terurut')
plt.legend()
sns.despine()
plt.show()
```
